```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

from common_plotting import read_data, plot, years, months, yearsFmt
```

```python
thp_data, data = read_data(['ks3458a-2w-res-log.csv', 'ks3458a-4w-res-log.csv', 'ks3458a1-2w-res-log.csv', 'ks3458a1-4w-res-log.csv'])
res_2w = data['ks3458a-2w-res-log.csv']
res_4w = data['ks3458a-4w-res-log.csv']
res_2w_1 = data['ks3458a1-2w-res-log.csv']
res_4w_1 = data['ks3458a1-4w-res-log.csv']
```

```python
display(res_2w['dut'].unique())
display(res_2w_1['dut'].unique())
display(res_4w['dut'].unique())
display(res_4w_1['dut'].unique())
```

```python
res_2w_groups = (res_2w[['dut', 'dut_setting']].apply(tuple, axis=1) != res_2w[['dut', 'dut_setting']].shift().apply(tuple, axis=1)).cumsum().rename('group')
res_2w_with_groups = res_2w.join(res_2w_groups)
res_2w_grouped = res_2w_with_groups.groupby('group').agg({'ag3458a_2_ohm': ['mean', 'std', 'sem', 'count'], 'temperature': ['mean', 'std', 'sem', 'count'], 'dut': 'last', 'dut_setting': 'last'})
res_4w_groups = (res_4w[['dut', 'dut_setting']].apply(tuple, axis=1) != res_4w[['dut', 'dut_setting']].shift().apply(tuple, axis=1)).cumsum().rename('group')
res_4w_with_groups = res_4w.join(res_4w_groups)
res_4w_grouped = res_4w_with_groups.groupby('group').agg({'ag3458a_2_ohm': ['mean', 'std', 'sem', 'count'], 'temperature': ['mean', 'std', 'sem', 'count'], 'dut': 'last', 'dut_setting': 'last'})
```

```python
display(res_2w_grouped)
display(res_4w_grouped)
```
